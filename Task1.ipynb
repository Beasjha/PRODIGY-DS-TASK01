```python
data_new=data.drop(['Country Code','Indicator Name','Indicator Code'], axis=1)
data_new.head(4)
```

```python
data_new.tail(4)
```

```python
x_country='India'
c_data=data_new[data_new['Country Name']==x_country]
year=c_data.columns[2:].astype(int) #used to convert datatype of array to integers 
population=c_data.iloc[:,2:].values.flatten() #convert the DataFrame into a 1-D array
plt.bar(year,population,color='pink')
plt.xlabel("Years")
plt.ylabel("India's Population")
plt.title('Population vs Years graph of India')
plt.show()
```

```python
plt.hist(population,bins=20, color= 'Orange',edgecolor='black')
plt.show()
```
